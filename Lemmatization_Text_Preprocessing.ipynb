```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')   # IMPORTANT: needed for newer NLTK versions
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS -Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("dancing",pos='v')
```

```python
words=["eating","eat","sing","singing","dance","dancing","finally","finalized","history"]
```

```python
for word in words:
    print(word+"----->"+lemmatizer.lemmatize(word))
```
