```python
corpus=""" Hello Welcome,I'm Anushika Jain.here I'm learning  Natural Language Processing.
In NLP today we'recreating some tokens.
"""
```

```python
corpus
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
##tokenization
##sentence-->paragraph

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentences in documents:
    print(sentences)
```

```python
## Tokenization
## Paragraph--> words
## sentence---> words
```

```python
 from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentences in documents:
  print(word_tokenize(sentences))
```

```python
from nltk import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
